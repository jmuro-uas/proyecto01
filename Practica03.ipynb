## 21.Limpieza y normalización de texto con pandas  
La **limpieza y normalización de texto con pandas** consiste en preparar datos textuales para que sean consistentes, analizables y útiles en procesos como análisis de datos, minería de texto, machine learning o ETL. En muchos conjuntos de datos, el texto viene con problemas como:

Mayúsculas y minúsculas mezcladas
Espacios innecesarios
Acentos
Caracteres especiales
Errores de escritura
Diferentes formas de escribir lo mismo
Valores nulos

Con pandas se utilizan principalmente las funciones de la serie .str para transformar y limpiar cadenas.

```python
import pandas as pd

df = pd.DataFrame({
    "nombre": ["  Juan PÉREZ ", "ana lópez", "CARLOS García  ", "  maría MARTÍNEZ"]
})

df["nombre_limpio"] = (df["nombre"]
    .str.strip()           # quitar espacios al inicio y final
    .str.lower()           # todo a minúsculas
    .str.title()           # Primera letra en mayúscula por palabra
)

print(df)
```

## 22. Eliminar caracteres especiales y símbolos

```python
df = pd.DataFrame({
    "descripcion": [
        "Servidor@Dell#2024!",
        "Switch   Cisco  24p",
        "Router//Mikrotik--RB",
        "UPS$APC***1500VA"
    ]
})

# Quitar todo excepto letras, números y espacios
df["desc_limpia"] = df["descripcion"].str.replace(r"[^a-zA-Z0-9\s]", "", regex=True)

# Colapsar espacios múltiples en uno solo
df["desc_limpia"] = df["desc_limpia"].str.replace(r"\s+", " ", regex=True).str.strip()

print(df)
```

## 23. Normalizar acentos y caracteres Unicode

```python
import unicodedata

def quitar_acentos(texto):
    if pd.isna(texto):
        return texto
    return unicodedata.normalize("NFD", texto)\
                      .encode("ascii", "ignore")\
                      .decode("utf-8")

df = pd.DataFrame({
    "ciudad": ["Culiacán", "Mérida", "Guadalajara", "Querétaro", "León"]
})

df["ciudad_sin_acento"] = df["ciudad"].apply(quitar_acentos)

print(df)
```

## 24. Estandarizar formatos: teléfonos

```python
df = pd.DataFrame({
    "telefono": [
        "667-123-4567",
        "(667) 123 4567",
        "6671234567",
        "+52 667 123 4567",
        "667.123.4567"
    ]
})

# Quedarse solo con dígitos
df["solo_digitos"] = df["telefono"].str.replace(r"\D", "", regex=True)

# Tomar los últimos 10 dígitos (quitar lada internacional)
df["telefono_std"] = df["solo_digitos"].str[-10:]

# Formato uniforme: XXX-XXX-XXXX
df["telefono_fmt"] = df["telefono_std"].str.replace(
    r"(\d{3})(\d{3})(\d{4})", r"\1-\2-\3", regex=True
)

print(df[["telefono", "telefono_fmt"]])
```

# 25. Estandarizar correos electrónicos

```python
df = pd.DataFrame({
    "correo": [
        "  Juan.Perez@UAS.EDU.MX  ",
        "ANA.LOPEZ@gmail.com",
        "carlos@EMPRESA.com.mx ",
        "MARIA@outlook.COM"
    ]
})

df["correo_std"] = (df["correo"]
    .str.strip()
    .str.lower()
)

# Validar formato básico con regex
df["es_valido"] = df["correo_std"].str.match(r"^[\w\.\-]+@[\w\-]+\.[a-z]{2,}(\.[a-z]{2,})?$")

print(df)
```

## 26. Estandarizar categorías con variantes

```python
df = pd.DataFrame({
    "departamento": [
        "Sistemas", "sistemas", "SISTEMAS", "Sis.",
        "Redes", "redes", "REDES",
        "Soporte", "soporte técnico", "Soporte Tecnico"
    ]
})

# Mapa de normalización
mapa = {
    r"sis.*":      "Sistemas",
    r"red.*":      "Redes",
    r"sop.*":      "Soporte",
}

def normalizar_dept(valor):
    valor_lower = str(valor).lower().strip()
    for patron, estandar in mapa.items():
        if pd.Series([valor_lower]).str.match(patron, case=False)[0]:
            return estandar
    return valor  # sin cambios si no hay coincidencia

df["dept_std"] = df["departamento"].apply(normalizar_dept)
print(df)
```

## 27. Extraer información con patrones

## 28. Pipeline completo de limpieza

```python
data = {
    "nombre":  ["Ana", "Luis", "   luis  ", " María7 ", "Pedro", "Ana", "sofía"],
    "edad":    [25, 32, 32, "veintiocho", 150, 25, np.nan],
    "salario": [15000, 22000, 22000, 18500, 999999, 15000, np.nan],
    "ciudad":  ["Culiacán", "Mazatlán", "Mazatlán", "culiacán", np.nan, "Culiacán", "Los Mochis"],
    "fecha":   ["2024-01-15", "2024/03/20", "2024/03/20", "15-04-2024", "2024-05-10", "2024-01-15", "2024-06-01"],
}

df = pd.DataFrame(data)

def limpiar_texto(serie: pd.Series) -> pd.Series:
    return (serie
        .astype(str)                                        # asegurar tipo string
        .str.strip()                                        # quitar espacios extremos
        .str.lower()                                        # minúsculas
        .str.replace(r"\s+", " ", regex=True)              # espacios múltiples
        .str.replace(r"[^a-záéíóúüñ\s0-9]", "", regex=True)  # solo letras/números
        .str.normalize("NFC")                               # normalizar unicode
    )

df["nombre_final"] = limpiar_texto(df["nombre"])
```

## Normalización min–max y z-score

Normalizar datos significa transformar los valores de una columna para que estén en una escala común, sin que pierdan su relación entre sí.
El problema que resuelve es este: si tienes una columna de edades (25–45) y otra de salarios (15,000–80,000), la columna de salarios domina cualquier cálculo simplemente por tener números más grandes, aunque no sea más importante. Normalizar pone ambas columnas en el mismo nivel.

## Min-Max — comprime todos los valores al rango [0, 1]:  
valor normalizado = (x - mínimo) / (máximo - mínimo)  
El valor más pequeño siempre queda en 0, el más grande en 1, y todos los demás proporcionales entre ellos.  
  
Salarios:   15,000 → 0.0  
          22,000 → 0.44  
          31,000 → 1.0
